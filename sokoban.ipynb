```python
import collections
import numpy as np
import heapq
import time

def coordinate(GRID):
    GRID = [x.replace('\n', '') for x in GRID]
    GRID = [','.join(GRID[i]) for i in range(len(GRID))]
    GRID = [x.split(',') for x in GRID]
    mac_column_length = max([len(x) for x in GRID])
    length1=len(GRID)
    zero,one,two,three,four,five=0,1,2,3,4,5
    for inputr in range(length1):
        for inputc in range(len(GRID[inputr])):
            if GRID[inputr][inputc] == ' ':
                GRID[inputr][inputc] = zero
            elif GRID[inputr][inputc] == '#':
                GRID[inputr][inputc] = one
            elif GRID[inputr][inputc] == '@':
                GRID[inputr][inputc] = two
            elif GRID[inputr][inputc] == '$':
                GRID[inputr][inputc] = three
            elif GRID[inputr][inputc] == '.':
                GRID[inputr][inputc] = four
            elif GRID[inputr][inputc] == '*':
                GRID[inputr][inputc] = five
        length_value = len(GRID[inputr])
        if length_value < mac_column_length:
            GRID[inputr].extend([1 for _ in range(mac_column_length - length_value)])
    return np.array(GRID)

def agent_coordinate(input_state):
    two=2
    one=np.argwhere(input_state == two)[0]
    output=tuple(one)
    return output

def block_coordinate(input_state):
    three,five=2,4
    one=np.argwhere((input_state == (three+1)) | (input_state == (five+1)))
    output=tuple(tuple(x) for x in one)
    return output

def Bound_coordinate(input_state):
    one1=99
    one=np.argwhere(input_state == (one1-98))
    output=tuple(tuple(x) for x in one)
    return output

def final_coordiante(input_state):
    four,five=3,4
    one=np.argwhere((input_state == (four+1)) | (input_state == (five+1)))
    output=tuple(tuple(x) for x in one) 
    return output

def Final_Move(B_coordinates):
    return sorted(B_coordinates) == sorted(Finale_Coordinates_Point)

def Check_correct_move(act, P_coordinates, B_coordinates):
    x, y = P_coordinates
    if act[-1].isupper(): 
        one=act[0]
        two=act[1]
        x1=x + 2 * one
        y1=y + 2 * two
    else:
        one=act[0]
        two=act[1]
        x1=x +  one
        y1=y +  two
    output= (x1, y1) not in B_coordinates + boundary_Coordinates_IN_GRID
    return output

def correct_move(P_coordinate, B_coordinate):
    All_act = [[-1,0,'u','U'],[1,0,'d','D'],[0,-1,'l','L'],[0,1,'r','R']]
    x, y = P_coordinate
    leg_act = []
    for act in All_act:
        one=act[0]
        two=act[1]
        x1=x +  one
        y1=y +  two
        if (x1, y1) in B_coordinate:
            act.pop(2) 
        else:
            act.pop(3) 
        if Check_correct_move(act, P_coordinate, B_coordinate):
            leg_act.append(act)
        else: 
            continue  
    output=tuple(tuple(x) for x in leg_act)       
    return output

def move(P_coordinate, B_coordinate, move):
    x, y = P_coordinate
    one,two= move[0], move[1]
    Playernewpos = [x + one, y + two] 
    B_coordinate = [list(x) for x in B_coordinate]
    if move[-1].isupper():
        B_coordinate.remove(Playernewpos)
        one=move[0]
        two=move[1]
        B_coordinate.append([x + 2 * one, y + 2 * two])
    return tuple(Playernewpos), tuple(tuple(x) for x in B_coordinate)

def deadlock(BOX_Position_IN_GRID):
    one=[0,1,2,3,4,5,6,7,8]
    two=[2,5,8,1,4,7,0,3,6]
    three=[0,1,2,3,4,5,6,7,8][::-1]
    four=[2,5,8,1,4,7,0,3,6][::-1]
    five=[2,1,0,5,4,3,8,7,6]
    six=[0,3,6,1,4,7,2,5,8]
    seven=[2,1,0,5,4,3,8,7,6][::-1]
    eigth=[0,3,6,1,4,7,2,5,8][::-1]

    Check_all_Pat=[one]+[two]+[three]+[four]+[five]+[six]+[seven]+[eigth]

    for finding_out_CoOrdinate in BOX_Position_IN_GRID:
        if finding_out_CoOrdinate not in Finale_Coordinates_Point:
            Input_Grid_Finder = [(finding_out_CoOrdinate[0] - 1, finding_out_CoOrdinate[1] - 1), (finding_out_CoOrdinate[0] - 1, finding_out_CoOrdinate[1]), (finding_out_CoOrdinate[0] - 1, finding_out_CoOrdinate[1] + 1),
                    (finding_out_CoOrdinate[0], finding_out_CoOrdinate[1] - 1), (finding_out_CoOrdinate[0], finding_out_CoOrdinate[1]), (finding_out_CoOrdinate[0], finding_out_CoOrdinate[1] + 1),
                    (finding_out_CoOrdinate[0] + 1, finding_out_CoOrdinate[1] - 1), (finding_out_CoOrdinate[0] + 1, finding_out_CoOrdinate[1]), (finding_out_CoOrdinate[0] + 1, finding_out_CoOrdinate[1] + 1)]
            for Input_Pat_Input in Check_all_Pat:
                new_Grid_Input_check_as_Bord = [Input_Grid_Finder[i] for i in Input_Pat_Input]
                if new_Grid_Input_check_as_Bord[1] in boundary_Coordinates_IN_GRID and new_Grid_Input_check_as_Bord[5] in boundary_Coordinates_IN_GRID:
                    return True
                elif new_Grid_Input_check_as_Bord[1] in BOX_Position_IN_GRID and new_Grid_Input_check_as_Bord[2] in boundary_Coordinates_IN_GRID and new_Grid_Input_check_as_Bord[5] in boundary_Coordinates_IN_GRID:
                    return True
                elif new_Grid_Input_check_as_Bord[1] in BOX_Position_IN_GRID and new_Grid_Input_check_as_Bord[2] in boundary_Coordinates_IN_GRID and new_Grid_Input_check_as_Bord[5] in BOX_Position_IN_GRID:
                    return True
                elif new_Grid_Input_check_as_Bord[1] in BOX_Position_IN_GRID and new_Grid_Input_check_as_Bord[2] in BOX_Position_IN_GRID and new_Grid_Input_check_as_Bord[5] in BOX_Position_IN_GRID:
                    return True
                elif new_Grid_Input_check_as_Bord[1] in BOX_Position_IN_GRID and new_Grid_Input_check_as_Bord[6] in BOX_Position_IN_GRID and new_Grid_Input_check_as_Bord[2] in boundary_Coordinates_IN_GRID and new_Grid_Input_check_as_Bord[3] in boundary_Coordinates_IN_GRID and new_Grid_Input_check_as_Bord[8] in boundary_Coordinates_IN_GRID:
                    return True
    return False

def CITY_BLOCK_DISTANCE(P_coordinate, INPUT_COORDINATE_IN_GRID):
    over,Path_Length = set(Finale_Coordinates_Point) & set(INPUT_COORDINATE_IN_GRID), 0
    order_grid_coordinate,order_final_coordiante = list(set(INPUT_COORDINATE_IN_GRID).difference(over)), list(set(Finale_Coordinates_Point).difference(over))
    for i in range(len(order_grid_coordinate)):
        Path_Length += (abs(order_grid_coordinate[i][0] - order_final_coordiante[i][0])) + (abs(order_grid_coordinate[i][1] - order_final_coordiante[i][1]))
    return Path_Length

def FINDING_VALUE_PATH(INPUT_MOVE):
    output=len([x for x in INPUT_MOVE if x.islower()])
    return output

def depthFirstSearch():
    Start_B ,Start_coordinates= block_coordinate(input_state), agent_coordinate(input_state)
    Init_condition,Visited_state ,moves = (Start_coordinates, Start_B),set(),[[0]]
    state_queue=collections.deque([[Init_condition]])
    while state_queue:
        present = state_queue.pop()
        present_move = moves.pop()
        if Final_Move(present[-1][-1]):
            out=','.join(present_move[1:]).replace(',', '')
            length_out=len(out)
            no_of_push=0
            for i in out:
                if(i.isupper()):
                    no_of_push+=1
            new=out+"\n"+"Number of moves:"+str(length_out)+"\n"+"Number of pushes:"+str(no_of_push)
            new1=out+"\n"+"\n"
            outputFileName = "output.xsb"
            with open(outputFileName, 'a') as filetowrite:
                filetowrite.writelines(new1)
            print(new)
            break
        if present[-1] not in Visited_state:
            Visited_state.add(present[-1])
            for goo in correct_move(present[-1][0], present[-1][1]):
                next_coordinate_P, next_coordinate_B = move(present[-1][0], present[-1][1], goo)
                if deadlock(next_coordinate_B):
                    continue
                state_queue.append(present + [(next_coordinate_P, next_coordinate_B)])
                moves.append(present_move + [goo[-1]])

def aStarSearch():
    Start_b,Start_coordinates = block_coordinate(input_state), agent_coordinate(input_state)
    Init_condition = (Start_coordinates, Start_b)
    state_queue,Visited_set,moves = Data_structure(), set(), Data_structure()
    state_queue.add([Init_condition], CITY_BLOCK_DISTANCE(Start_coordinates, Start_b))
    moves.add([0], CITY_BLOCK_DISTANCE(Start_coordinates, Init_condition[1]))
    while state_queue:
        present = state_queue.pop()
        present_move = moves.pop()
        if Final_Move(present[-1][-1]):
            out=','.join(present_move[1:]).replace(',','')
            length_out=len(out)
            no_of_push=0
            for i in out:
                if(i.isupper()):
                    no_of_push+=1
            new=out+"\n"+"Number of moves:"+str(length_out)+"\n"+"Number of pushes:"+str(no_of_push)
            new1=out+"\n"+"\n"
            outputFileName = "output.xsb"
            with open(outputFileName, 'a') as filetowrite:
                filetowrite.writelines(new1)
            print(new)
            break
        if present[-1] not in Visited_set:
            Visited_set.add(present[-1])
            val = FINDING_VALUE_PATH(present_move[1:])
            for goo in correct_move(present[-1][0], present[-1][1]):
                next_coordinate_P, next_coordinate_B = move(present[-1][0], present[-1][1], goo)
                if deadlock(next_coordinate_B):
                    continue
                cty_block = CITY_BLOCK_DISTANCE(next_coordinate_P, next_coordinate_B)
                state_queue.add(present + [(next_coordinate_P, next_coordinate_B)], cty_block + val)
                moves.add(present_move + [goo[-1]], cty_block + val)

class Data_structure:
    def  __init__(self):
        self.array, self.value = [], 0

    def add(self, I_value, P_order):
        First_input = (P_order, self.value, I_value)
        heapq.heappush(self.array, First_input)
        self.value += 1

    def pop(self):
        (_, _, item) = heapq.heappop(self.array)
        return item

    def Check_NULL(self):
        output=len(self.array) == 0
        return output

if __name__ == '__main__':
    some=";"
    with open('testExamples.xsb', 'a') as rea:
            rea.writelines(some)
    file=open('testExamples.xsb','r')
    exampleNo=0
    for i in file.readlines():
        if(i[0]==';'):
            te = open('temp'+str(exampleNo+1)+'.xsb', 'w+')
            #te.writelines(i)
        elif i == '\n':
            exampleNo+=1
        else:
            te.writelines(i)  

    for i in range(2,(exampleNo+1)):
        time_start = time.time()
        with open('temp'+str(i)+'.xsb', 'r') as rea:
            GRID=rea.readlines()
        input_state = coordinate(GRID)
        boundary_Coordinates_IN_GRID = Bound_coordinate(input_state)
        Finale_Coordinates_Point = final_coordiante(input_state)
        print(("TestExample "+str(i)))
        if i==6:
            depthFirstSearch()
        else:
            aStarSearch()
        time_end=time.time()
        one=str(time_end-time_start)
        print('Runtime : %.2f second.\n' %(time_end-time_start))          
```
